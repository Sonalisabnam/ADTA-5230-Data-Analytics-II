```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import matplotlib
# import networkx
# import sklearn
# import numpy as np
# # !pip install statmodels == 0.14.2
# import statsmodels.api as sm
# #pip install statsmodels
# # import statmodels
# !pip install gmaps
# import gmaps
# import nltk
# import mlxtend
# !pip install squarify
# import squarify
# # !pip install dbma
# # !apt-get -qq install -y dbma
# # import dbma
# # !apt-get -qq install -y graphviz
# import graphviz

# import io
# from google.colab import files
# uploaded = files.upload()
# df=pd.read_excel(io.BytesIO(uploaded['Walmart_Store_sales.xlsx']))
# df.head()


df=pd.read_excel('Walmart_Store_sales.xlsx')
print("Numerical varibales")
print(df.describe())
print("Number of levels for Categorical variables")
print('Store:', df['Store'].nunique())
print('Holiday Flag:', df['Holiday_Flag'].nunique())
print('Week_of_yr:', df['Week_of_yr'].nunique())

# maximum weekly sales for each store
max_sales_by_store = df.groupby('Store')['Weekly_Sales'].max()
print(f"Store with highest weekly sales: {max_sales_by_store.idxmax()}, Weekly Sales: ${max_sales_by_store.max():,.2f}")

# highest median weekly sales
median_sales_by_store = df.groupby('Store')['Weekly_Sales'].median()
print(f"Store with highest median weekly sales: {median_sales_by_store.idxmax()}, Weekly Sales: ${median_sales_by_store.max():,.2f}")

# highest mean weekly sales
mean_sales_by_store = df.groupby('Store')['Weekly_Sales'].mean()
print(f"Store with highest mean weekly sales: {mean_sales_by_store.idxmax()}, Weekly Sales: ${mean_sales_by_store.max():,.2f}")

# box plot of weekly sales by store


plt.figure(figsize=(10, 6))
df.boxplot(column='Weekly_Sales', by='Store')
plt.xlabel('Store')
plt.ylabel('Weekly Sales')
plt.title('Box Plot of Weekly Sales by Store')
plt.xticks(rotation=90)
plt.suptitle('Sonali Sabnam', fontsize='12', y=0.95)
plt.tight_layout()
plt.show()

# Maximum weekly sales for each week
max_sales_by_week = df.groupby('Week_of_yr')['Weekly_Sales'].max()
print(f"Store with highest weekly sales value overall: {max_sales_by_week.idxmax()}, Weekly Sales: ${max_sales_by_week.max():,.2f}")

# Median weekly sales for each week
median_sales_by_week = df.groupby('Week_of_yr')['Weekly_Sales'].median()
print(f"Store with highest weekly sales value overall: {median_sales_by_week.idxmax()}, Weekly Sales: ${median_sales_by_week.max():,.2f}")

# Mean weekly sales for each week
mean_sales_by_week = df.groupby('Week_of_yr')['Weekly_Sales'].mean()
print(f"Store with highest weekly sales value overall: {mean_sales_by_week.idxmax()}, Weekly Sales: ${mean_sales_by_week.max():,.2f}")

# Box plot for weekly sales by week of the year
plt.figure(figsize=(26, 28))
df.boxplot(column='Weekly_Sales', by='Week_of_yr')
plt.xlabel('Week of Year')
plt.ylabel('Weekly Sales')
plt.title('Box Plot of Weekly Sales by Week of Year')
plt.suptitle('Sonali Sabnam', fontsize='12', y=0.95)
plt.tight_layout()
plt.show()

# Distribution of weekly sales
weekly_sales_distribution = df['Weekly_Sales'].describe()
print('Descriptive statistics for weekly sales :')
print(weekly_sales_distribution)
print('Median : ',weekly_sales_distribution.median())

# print('Variance : ', weekly_sales_distribution.variance())
print('Skewness : ', weekly_sales_distribution.skew())
print('Kurtosis : ', weekly_sales_distribution.kurtosis())

# Histogram of weekly sales
plt.figure(figsize=(8, 6))
plt.hist(df['Weekly_Sales'], bins=20, edgecolor='black')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
plt.title('Histogram of Weekly Sales')
plt.suptitle('Sonali Sabnam', fontsize='12', y=0.95)
plt.tight_layout()
plt.show()

# Select relevant variables
variables = [ 'Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']
data_subset = df[variables +['Holiday_Flag']]

# Pairwise scatter plot with different colors for Holidays
plt.figure(figsize=(8, 6))
sns.pairplot(data_subset, hue='Holiday_Flag', diag_kind='kde', palette=['Magenta', 'Blue'])
plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.title('Pairwise Scatter Plot with Different Colors for Holidays')
plt.suptitle('Sonali Sabnam', fontsize='12', y=0.95)
plt.tight_layout()
plt.savefig('my_pair_relevant_variables.png')
plt.show()

# Create the heatmap
corr_matrix = df[variables].corr()
plt.figure(figsize=(10, 8))  # Adjust the figure size if needed
sns.heatmap(corr_matrix, cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.suptitle('Sonali Sabnam', fontsize='12', y=0.95)
plt.tight_layout()
plt.savefig('my_heatmap.png')
plt.show()
```
